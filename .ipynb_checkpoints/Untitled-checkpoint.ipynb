```python
identity = [['Basketball','Swimming', 'Softball','Volleyball'],
           "Female", "D1"]
football = ["Football","Male",["HS1","HS2","D1","D3"]]
otherSport = [["Baseball","Basketball"], "Male", "D1"]
```

```python
def female(val):
    for i in range(len(identity[0])):
        BF1= df[(df["Sport"]==identity[0][i])&
                (df["Gender"]==identity[1])&
                (df["School"]==identity[2])].reset_index(drop=True)
        num = 0
        num2= 0
        Bench = BF1.columns[5]
        Squat = BF1.columns[6]
        Gender = BF1.Gender.unique()[0]
        Sport = BF1.Sport.unique()[0]+" Players"
        if Sport == "Swimming Players":
            Sport = "Swimmers"
        School = BF1.School.unique()[0]
        if val[0]>BF1.Bench[0]:
            print("Your Bench({}) is over the 90th Percentile for {} {} {}. Lay off the Steroids\n".format(val[0],
                                                                                                          School,Gender,Sport))
        else:
            for i in BF1.Bench:
                if val[0] <= i:
                    num+=1
            if num ==9:
                percentile = 0
            else:
                percentile = BF1.Percentile[num]
            
            print(("Your {}({}) is in the {}th percentile for {} {} {}\n").format(Bench,val[0],percentile,School,Gender,Sport))
            
        if val[1]>BF1.Squat[0]:
            print("Your Squat({}) is over the 90th Percentile for {} {} {}. Lay off the Steroids\n".format(val[1],
                                                                                                          School,Gender,Sport))
        else:
            for i in BF1.Squat:
                if val[1] <= i:
                    num2+=1
            if num2 ==9:
                percentile = 0
            else:
                percentile = BF1.Percentile[num2]
            
            print(("Your {}({}) is in the {}th percentile for {} {} {} Players\n").format(Squat,
                                                                                         val[1],percentile,School,Gender,Sport))
```

```python
def other(val):
    for i in range(len(otherSport[0])):
        BF1= df[(df["Sport"]==otherSport[0][i])&
                (df["Gender"]==otherSport[1])&
                (df["School"]==otherSport[2])].reset_index(drop=True)
        num = 0
        num2 = 0
        Bench = BF1.columns[5]
        Squat = BF1.columns[6]
        Gender = BF1.Gender.unique()[0]
        Sport = BF1.Sport.unique()[0]
        School = BF1.School.unique()[0]
        if val[0]>BF1.Bench[0]:
            print("Your Bench({}) is over the 90th Percentile for {} {} {}. Lay off the Steroids\n".format(val[0],
                                                                                                          School,Gender,Sport))
        
        else:
            for i in BF1.Bench:
                if val[0] <= i:
                    num+=1
            if num ==9:
                percentile = 0
            else:
                percentile = BF1.Percentile[num]
            
            print(("Your {}({}) is in the {}th percentile for {} {} {}\n").format(Bench,val[0],percentile,School,Gender,Sport))
        if val[1]>BF1.Squat[0]:
            print("Your Squat ({}) is over the 90th Percentile for {} {} {}. Lay off the Steroids\n".format(val[1],
                                                                                                          School,Gender,Sport))
        else:
            for i in BF1.Squat:
                if val[1] <= i:
                    num2+=1
            if num2 ==9:
                percentile = 0
            else:
                percentile = BF1.Percentile[num2]
            
            print(("Your {}({}) is in the {}th percentile for {} {} {} Players\n").format(Squat,
                                                                                         val[1],percentile,School,Gender,Sport))
```

```python
def fball(val):
    for i in range(len(football[2])):
        BF1= df[(df["Sport"]==football[0])&
                (df["Gender"]==football[1])&
                (df["School"]==football[2][i])].reset_index(drop=True)
        num = 0
        num2 = 0
        Bench = BF1.columns[5]
        Squat = BF1.columns[6]
        Gender = BF1.Gender.unique()[0]
        Sport = BF1.Sport.unique()[0]
        School = BF1.School.unique()[0]
        if School == "HS1":
            School = "High School(ages 14-16)"
        elif School == "HS2":
            School = "High School(ages 16-18)"
        else:
            School = School
        if val[0]>BF1.Bench[0]:
            print("Your Bench({}) is over the 90th Percentile for {} {} {}. Lay off the Steroids\n".format(val[0],
                                                                                                          School,Gender,Sport))
        else:
            for i in BF1.Bench:
                if val[0] <= i:
                    num+=1
            if num ==9:
                percentile = 0
            else:
                percentile = BF1.Percentile[num]
            
            print(("Your {}({}) is in the {}th percentile for {} {} {}\n").format(Bench,val[0],percentile,School,Gender,Sport))
        if val[1]>BF1.Squat[0]:
            print("Your Squat({}) is over the 90th Percentile for {} {} {}. Lay off the Steroids\n".format(val[1],
                                                                                                          School,Gender,Sport))
        else:
            for i in BF1.Squat:
                if val[1] <= i:
                    num2+=1
            if num2 ==9:
                percentile = 0
            else:
                percentile = BF1.Percentile[num2]
            
            print(("Your {}({}) is in the {}th percentile for {} {} {} Players\n").format(Squat,
                                                                                         val[1],percentile,School,Gender,Sport))
```

```python
def values():
    while True:
        try:
            bench = int(input("Enter Your Bench "))
            break
        except:
            print('Enter a number please')
    while True:
        try:
            squat = int(input("Enter Your Squat "))
            break
        except:
            print('Enter a number please')
    return [bench,squat]
```

```python
def main():
    lifts = values()
    print("\nYour Rank")
    section = "\n-----------------------------------------------------------------------------------\n"
    print(section)
    print("Female Lifts")
    female(lifts)
    print(section)
    print("Male Lifts\n")
    other(lifts)
    print(section)
    print("Football\n")
    fball(lifts)
```
